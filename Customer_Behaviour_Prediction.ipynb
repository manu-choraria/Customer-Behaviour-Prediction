```python
#Import the csv file
import pandas as pd
file = pd.read_csv("/content/Customer_Behaviour.csv")

#print Dataset
print(file)
```

```python
#print first five data
file.head()
```

```python
#print last five data
file.tail()
```

```python
#print information about dataset
file.info()
```

```python
#five point summary, mean and standard deviation of dataset
file.describe()
```

```python
#To print missing values and number of missing values
print(file.isnull())
print('Total number of missing values per column:')
file.isnull().sum()
```

```python
#To convert Male-Female into numeric values
file=pd.get_dummies(file,drop_first=True) #Here drop first indicated that Gender_female will be deleted
file.head(5)
```

```python
#To fill missing values using KNN
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
file = pd.DataFrame(imputer.fit_transform(file),columns = file.columns)
print(file)
print('Total number of missing values per column:')
file.isnull().sum()
```

```python
#To mark labels
import numpy as np
labels = np.array(file['Purchased'])

#To get only features
features= file.drop('Purchased', axis = 1)

#To convert features into array
features = np.array(features)

#split data in training and testing datasets
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
#To train model using Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
gaussian_model = GaussianNB()
gaussian_model.fit(train_features, train_labels)

#To test model
y_pred = gaussian_model.predict(test_features)

#To measure accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test_labels, y_pred)*100
print("Accuracy of Gaussian Model is "+str(accuracy)+"%")
```

```python
#To train model using Logistic Regression
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()
logistic_model.fit(train_features, train_labels)

acc = logistic_model.score(test_features,test_labels)

print("Accuracy of Logistic Regression Model is "+str(acc*100)+"%")
```
